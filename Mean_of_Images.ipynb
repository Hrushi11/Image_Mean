```python
# Generating an array of all images
all_imgs = []
for elem in num:
  name = dir + elem + ".jpg"
  img = plt.imread(name)
  img = img.flatten()
  all_imgs.append(img)
  
all_imgs = np.array(all_imgs)
```

Now we are trying to do this: 

![IMG](https://cdn.hashnode.com/res/hashnode/image/upload/v1619245596430/YXlRKGbxz.jpeg?auto=compress,format&format=webp)

```python
# Taking the transpose
all_imgs_trans = all_imgs.T
all_imgs_trans
```

```python
def mean_of_img(all_imgs):
  assert len(all_imgs[0]) == len(all_imgs[1])
  mean_img = []
  for img_num in range(len(all_imgs)):
    mean_img.append(np.mean(all_imgs[img_num]))
  mean_img = np.array(mean_img)
  
  return mean_img
```

```python
# 28 * 28
mean_img = mean_of_img(all_imgs_trans)
len(mean_img)
```

```python
# checking the array
mean_img
```

```python
mean_img = mean_img.reshape((28, 28))
len(mean_img)
```

```python
# Plotting the mean image
plt.title("Mean of Images", color="green")
plt.axis(False)
plt.imshow(mean_img);
```

The images as you can see in this case are not of the same sizes so the flattened array will be of different sizes. The best way to solve this is to resize all the images to one format say `(128 x 128)`. 

```python
# To resize the image
def resize(img):
  img = tf.image.resize(img, (128, 128))

  return np.array(img)
```

```python
img = plt.imread(dir + "1.jpg")
print(f"Size of original image: {img.shape} = {img.shape[0] * img.shape[0] * img.shape[0]}")
img = resize(img)
print(f"Size of resized image: {img.shape} = {img.shape[0] * img.shape[0] * img.shape[0]}")
```

```python
# Generating an array of all images
all_imgs = []
for elem in img_nums:
  name = dir + str(elem) + ".jpg"
  img = plt.imread(name)
  img = resize(img)
  img = img.flatten()
  all_imgs.append(img)
  
all_imgs = np.array(all_imgs)
```

```python
def mean_of_img(all_imgs_trans):
  assert len(all_imgs_trans[0]) == len(all_imgs_trans[1])
  mean_img = []
  for img_num in range(len(all_imgs_trans)):
    mean_img.append(np.mean(all_imgs_trans[img_num]))
  mean_img = np.array(mean_img)
  
  return mean_img
```
