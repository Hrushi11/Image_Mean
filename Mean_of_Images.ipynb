## Function

Creating a single function that just takes the path to the directory and finds the mean image.

```python
# Single function to generate the mean
def mean_gen(directory, shape=(128, 128)):
  """
  directory -> path to the directory that contains the images
  shape -> shape to be considered while taking the mean
  ---
  returns -> img (Mean of all images in the the directory) 
  """
  # To store flattened images
  all_imgs = []

  # To get all the images from the directory
  file_names = os.listdir(directory)

  # Looping over all images to resize and flatten them
  for i in range(len(file_names)):
    img = plt.imread(os.path.join(directory, file_names[i]))
    img = np.array(tf.image.resize(img, shape))
    img = img.flatten()
    all_imgs.append(img)
  
  # Transposing the all images matrix
  all_imgs_trans = np.array(all_imgs).T

  # Making sure that images are of same size
  assert len(all_imgs_trans[0]) == len(all_imgs_trans[-1])

  # To store the mean values and then turn this matrix to image
  mean_img = []
  
  # Calculating the mean of each row and appending in the mean image array
  for img_num in range(len(all_imgs_trans)):
    mean_img.append(np.mean(all_imgs_trans[img_num]))
  mean_img = np.array(mean_img)

  # Turning the flattened mean image into original image of shape (shape[0], shape[1], 3)
  mean_img = mean_img.reshape((shape[0], shape[1], 3))

  return mean_img
```

```python
# Plotting our dataset for which we will be finding the mean image
dir = "/content/Mean_In_Images/Faces/Women"
filenames = os.listdir(dir)
plt.figure(figsize=(17, 12))
for i in range(len(filenames)):
  img = plt.imread(os.path.join(dir, filenames[i]))
  img = tf.image.resize(img, (128, 128))
  title_ = filenames[i]

  # Plotting the images
  plt.subplot(3, 4, i+1)
  plt.imshow(img/255)
  plt.axis(False)
  plt.title(title_, color="blue");  
```

```python
# Plotting the mean image
mean_img = mean_gen(dir)
plt.imshow(mean_img/255)
plt.axis(False)
plt.title("Mean of women", color="green");
```
